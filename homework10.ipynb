```python
import sqlite3
import pandas as pd
con = sqlite3.connect("sakila.db")
```

```python
sql_query = '''SELECT * FROM payment WHERE payment_date BETWEEN '2002-02-14 15:16:03.000' AND '2005-11-10 00:00:00' '''
```

```python
df = pd.read_sql(sql_query, con)

df
```

```python
sql_query2 = '''SELECT * FROM payment WHERE payment_date >= '2002-02-14 15:16:03.000' OR payment_date <= '2005-11-10 00:00:00' '''
```

```python
df2 = pd.read_sql(sql_query2, con)

df2
```
