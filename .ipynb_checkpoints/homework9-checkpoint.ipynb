manager_staff_id上位2名の会計したお客最新20件

```python
import sqlite3
import pandas as pd
con = sqlite3.connect("sakila.db")
```

```python
sql_query='''SELECT A.staff_id, A.payment_date, A.customer_id, C.first_name, C.email, B.staff_id, B.payment_date, B.customer_id, D.first_name, D.email FROM payment AS A INNER JOIN payment AS B CROSS JOIN customer AS C INNER JOIN customer AS D ON A.customer_id = C.customer_id AND B.customer_id = D.customer_id WHERE A.staff_id = 1 AND B.staff_id = 2 ORDER BY A.payment_date DESC, B.payment_date DESC LIMIT 20'''
```

```python
df = pd.read_sql(sql_query, con)

df
```
