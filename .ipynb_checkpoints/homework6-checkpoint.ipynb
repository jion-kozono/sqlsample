副問合せ
国ごとcity_id がmaxなやつ。(保留)
<p>
例）
select *
from hogehoge
where city_id = ( select city_id from coutry )</p>

```python
import sqlite3
import pandas as pd
con = sqlite3.connect("sakila.db")
```

副問合せではないがこれが簡単↓

```python
sql_query = '''SELECT country, city, MAX(city.city_id) FROM country LEFT OUTER JOIN city ON country.country_id = city.country_id GROUP BY country'''
```

```python
df = pd.read_sql(sql_query, con)

df
```

うまく取得できない↓

```python
sql_query2='''SELECT country, city, MAX(city.city_id) FROM city CROSS JOIN country ON city.country_id = country.country_id WHERE country IN (SELECT country FROM country GROUP BY country)'''
```

```python
df2 = pd.read_sql(sql_query2, con)

df2
```

これが最適？↓

```python
sql_query3='''SELECT * FROM country CROSS JOIN(SELECT country, MAX(city.city_id) FROM country CROSS JOIN city ON country.country_id = city.country_id GROUP BY country)'''
```

```python
df3 = pd.read_sql(sql_query3, con)

df3
```
